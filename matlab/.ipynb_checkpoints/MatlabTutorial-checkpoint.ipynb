### Loading the data

Let's load the neural data we want to work with. There are several ways to save and load data in Matlab, here, we are reading the data from the same '.pkl' file that we used in the Python tutorial.

```matlab
data = loadData(1, 1);

% What does the data structure have in it?
disp(data)

% You can also explore data by double clicking on it in the Variable
% WorkSpace (with the Matlab GUI)
```

### Visualizing the data

In Matlab, we can visualize the data using the `plot` function. We do not need to import any libraries to use this function (as with most Matlab functions you will use - except [toolboxes like this](https://www.mathworks.com/products/signal.html)).

```matlab
% Generate the time vector as x, and the average ECoG data as y
x = (1:size(data.ecog, 2))/data.ecog_sr-0.5;
y = squeeze(mean(data.ecog, 1));
colors = [0.96 0.44 0.54; 0.74 0.60 0.19];

figure('Position', [100, 100, 800, 300]);
for el = 1:2
    plot(x, y(:, el), 'LineWidth', 2, 'Color', colors(el, :)); 
    hold on;
end

% Format the plot
xlabel('Time (s)');
ylabel('Average electrical activity');
box off;
set(gca, 'FontSize', 15);
```

### Run PCA on the data

Just like in Python, we can run PCA on the data to visualize the first two principal components. We will use the `pca` function in Matlab to do this.

```matlab
% First, we need to reshape the data
X = cat(1, data.ecog(:, :, 1), data.ecog(:, :, 2));

% What shape is X?
disp(size(X))
```

```matlab
% Run PCA
[coeff, score, latent, ~, explained] = pca(X);

% Plot the trials in the first two principal components
figure('Position', [100, 100, 150, 150]);
for el = 1:2
    scatter(score((el-1)*size(data.ecog, 1)+1:el*size(data.ecog, 1), 1), ...
        score((el-1)*size(data.ecog, 1)+1:el*size(data.ecog, 1), 2), ...
        30, colors(el, :), 'filled'); hold on;
end
hold off;

xlabel('PC1');
ylabel('PC2');
```
